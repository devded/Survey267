##### count mental_health_consequence 

```python
mental_health_consequence_yes = dataset[(dataset['mental_health_consequence']=="Yes")].shape[0]
mental_health_consequence_no = dataset[(dataset['mental_health_consequence']=="No")].shape[0]

print("Mental Health Consequence Yes :", mental_health_consequence_yes)
print("Mental Health Consequence No :", mental_health_consequence_no)
```

 ##### count seek_help based on suffering mental_health_consequence

```python
# seek_help (Yes, No) -> mental_health_consequence -> (Yes)
seek_help_yes = dataset[(dataset['seek_help']=="Yes") & (dataset['mental_health_consequence']=="Yes")].shape[0]
seek_help_no = dataset[(dataset['seek_help']=="No") & (dataset['mental_health_consequence']=="Yes")].shape[0]

print("Seek Help Yes : ", seek_help_yes)
print("Seek Help No : ", seek_help_no)
```

#### calculate seek_help perentage who is suffering mental_health_consequence

```python
seek_help_yes_percentage = (seek_help_yes * 100) / mental_health_consequence_yes
seek_help_no_percentage = (seek_help_no * 100) / mental_health_consequence_yes

print("Seek Help Yes Percentage : ", int(seek_help_yes_percentage) , "%")
print("Seek Help No Percentage : ", int(seek_help_no_percentage ) , "%")
```

##### count treatment  based on ask for seek_help and suffering mental_health_consequence

```python
# treatment (Yes, No) -> mental_health_consequence -> (Yes) -> seek_help -> (Yes)

treatment_yes = dataset[(dataset['treatment']=="Yes") & (dataset['seek_help']=="Yes") & (dataset['mental_health_consequence']=="Yes") ].shape[0]
treatment_no = dataset[(dataset['treatment']=="No") & (dataset['seek_help']=="Yes" ) & (dataset['mental_health_consequence']=="Yes")].shape[0]

print("Treatment Yes : ", treatment_yes)
print("Treatment No : ", treatment_no)
```

#### calculate treatment percentage

```python
treatment_yes_precentage = (treatment_yes*100)/ seek_help_yes
treatment_no_precentage = (treatment_no*100)/ seek_help_yes

print("Treatment Yes Percentage : ", int(treatment_yes_precentage) , "%")
print("Treatment No Percentage : ", int(treatment_no_precentage) , "%")
```
